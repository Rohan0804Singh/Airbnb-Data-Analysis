1) Find the distribution of Airbnb Bookings Price Range.

```python
sns.boxplot(df,x='price')
plt.show()
```

```python
#Removing outliers
```

```python
def outlier(column):
    percentile25=df[column].quantile(0.25)
    percentile75=df[column].quantile(0.75)
    iqr=percentile75-percentile25
    lower_range=percentile25-1.5*iqr
    upper_range=percentile75+1.5*iqr
    return lower_range,upper_range
```

```python
outlier('price')
```

```python
df.shape
```

```python
df1=df[(df['price']>-90) &(df['price']<334)]
```

```python
df1.shape
```

```python
#For distribution we will plot a histogram
sns.histplot(df1['price'],bins=20,kde=True)
plt.title("Distribution of Airbnb Bookings Price Range")
y=df1['price']
plt.show()
```
