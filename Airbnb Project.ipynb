```python
#Creating a function for removing outliers
def outlier(column):
    percentile25=df['price'].quantile(0.25)
    percentile75=df['price'].quantile(0.75)
    iqr=percentile75-percentile25
    upper_range=percentile75+1.5*iqr
    lower_range=percentile25-1.5*iqr
    return lower_range,upper_range
```

```python
outlier('price')
```
